```python
import pandas as pd
import os
import numpy as np

if os.path.isdir("/Users/morpholino/OwnCloud/"):
    homedir = "/Users/morpholino/OwnCloud/"
elif os.path.isdir("/Volumes/zoliq data/OwnCloud/"):
    homedir = "/Volumes/zoliq data/OwnCloud/"
else:
    print("Please set a homedir")
wd = homedir + "Jankoviny/Tick_transcriptome/monstrtabulka"
os.chdir(wd)


#INFILES, OUTFILE
array = {}
files = [f for f in os.listdir() if f.endswith("cols_sub.txt")] #zde nahradit "cols.txt"
indices = [f.replace("Msig.del.cols_sub.txt", "") for f in files] #odvozene z nazvu souboru, 
                                                                  #nahradit za "Msig.del.cols.txt"
indices.sort()
outtable = "final_table.tsv"

#MAIN
print("loading data...")
for file in files:
    with open(file) as f: 
        c = 0
        for l in f:
            if c == 0: #extract column names, which is the third line
                c += 1
                allcolumns = l.strip().split("\t")
                allcolumns = [x.replace('"','') for x in allcolumns]
                print("identified columns:")
                print(allcolumns)
            else: #process data below the column names line
                l = l.replace('"','')
                data = l.strip().split("\t")
                array[data[1]] = ["-", "-", "-", "-", "-", "-", "-", "-"] #eight undefs
print("reading unique seqIDs finished")

df = pd.DataFrame(array, index=indices, dtype=np.float64)

X = df.T
```

```python
print("continuing filling the dataframe")
for file in files:
    stage = file.replace("Msig.del.cols_sub.txt", "")
    print("analyzing file: ", file)
    with open(file) as f:     
        c = 0
        for l in f:
            if c == 0: #extract column names, which is the third line
                c += 1
                #print("header>skipping")
            else: #process data below the column names line
                l = l.replace('"','')
                #print(l)
                line = l.strip().split("\t")
                geneID, value = line[1], line[0]
                #print("geneID, value obtained")
                X.at[geneID, stage] = value
```
