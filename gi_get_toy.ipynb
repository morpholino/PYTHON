```python
#if accession is a gene, use ET as Entrez XML parser fails big time
import xml.etree.ElementTree as ET

accession = "PF3D7_0811900"
handle = Entrez.efetch(db="gene", id=accession, rettype="fasta", retmode="XML")
#weird thing, i got a human gene with this accession, I simply switched to regular fasta name search
record = ET.parse(handle)
print(record)
root = record.getroot()
root.tag
```

```python
for child in root.iter():
    try:
        print(child.text, child.attrib)
    except AttributeError:
        print(child.tag, "no attribute")
```

```python
root[0]["Prot-ref"]
```
